# 🏃‍♂️ Workout Data Exploration
This notebook allows you to directly query the CSV data stored in the `data/` directory.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Set paths
DATA_DIR = Path('data')

# Load data
df_activities = pd.read_csv(DATA_DIR / 'strava_activities.csv')
df_splits = pd.read_csv(DATA_DIR / 'strava_splits.csv')

print(f"Loaded {len(df_activities)} activities and {len(df_splits)} split records.")
```

## Quick Summary Stats

```python
# Top 5 most frequent activity types
df_activities['type'].value_counts().head()
```

```python
# Basic distance/pace stats for Runs
runs = df_activities[df_activities['type'] == 'Run'].copy()
# Note: distance is in meters in raw Strava data, the app adds distance_miles
if 'distance_miles' not in runs.columns:
    runs['distance_miles'] = runs['distance'] / 1609.34

runs[['distance_miles', 'average_heartrate', 'max_heartrate']].describe()
```

## Visualize Mileage Trends

```python
plt.figure(figsize=(12, 6))
runs['date'] = pd.to_datetime(runs['start_date']).dt.date
daily_miles = runs.groupby('date')['distance_miles'].sum()
daily_miles.rolling(window=7).mean().plot(label='7-Day Avg', color='#38bdf8')
plt.title('Daily Running Mileage (Trend)')
plt.ylabel('Miles')
plt.grid(alpha=0.2)
plt.legend()
plt.show()
```

## Efficiency: Pace vs Heart Rate

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=runs, x='distance_miles', y='average_heartrate', hue='type', alpha=0.6)
plt.title('Workout Intensity (Distance vs HR)')
plt.show()
```
